```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
import random
import string

schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("country", StringType(), True),
    StructField("email", StringType(), True),
    StructField("gender", StringType(), True),
    StructField("salary", IntegerType(), True),
    StructField("department", StringType(), True),
    StructField("city", StringType(), True),
    StructField("phone", StringType(), True)
])

def random_string(length=7):
    return ''.join(random.choices(string.ascii_lowercase, k=length))

def random_email(name):
    return f"{name}@example.com"

def random_phone():
    return ''.join(random.choices(string.digits, k=10))

def random_row(id):
    name = random_string()
    return (
        id,                          
        name.capitalize(),           
        random.randint(20, 60),      
        random.choice(["USA", "UK", "Canada", "Germany", "Australia"]), 
        random_email(name),           
        random.choice(["Male", "Female"]), 
        random.randint(30000, 150000),
        random.choice(["Sales", "IT", "HR", "Finance", "Marketing"]),
        random.choice(["New York", "London", "Toronto", "Berlin", "Sydney"]),
        random_phone()
    )

data = [random_row(i) for i in range(1, 50001)]

df = spark.createDataFrame(data, schema)
df.write.mode("overwrite").saveAsTable('customer')
```
